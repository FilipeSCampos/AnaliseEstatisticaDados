```python
freq_observadas = [1148, 275, 355, 450, 545, 640, 734, 828, 925]

total_observacoes = sum(freq_observadas)

num_digitos = len(freq_observadas)

prob_uniforme = 1 / num_digitos

probabilidades_esperadas = [prob_uniforme] * num_digitos

freq_esperadas = [prob * total_observacoes for  prob in probabilidades_esperadas]

prob_observadas = [freq / total_observacoes for freq in freq_observadas]

z = [(obs - esp) / math.sqrt(esp * (1 - prob)) for obs, esp, prob in zip(freq_observadas, freq_esperadas, probabilidades_esperadas)]
```

```python
for i in range(len(freq_observadas)):
    print(f"Dígito {i}:")
    print(f"  Frequência Observada: {freq_observadas[i]}")
    print(f"  Frequência Esperada: {freq_esperadas[i]}")
    print(f"  Probabilidade Observada: {prob_observadas[i]}")
    print(f"  Probabilidade Esperada: {probabilidades_esperadas[i]}")
    print(f"  Estatística Z: {z[i]}")
    print("\n")
```

```python
import pandas as pd
import math

# Defina os valores de frequência observada a partir da Tabela 2
frequencia_observada = [148, 75, 55, 50, 45, 40, 34, 28, 25]

# Número total de observações
n = 500

# Crie um DataFrame
df = pd.DataFrame({
    'Primeiro Dígito': [1, 2, 3, 4, 5, 6, 7, 8, 9],
    'Frequência observada': frequencia_observada
})

# Calcule as probabilidades esperadas usando a fórmula Pe(d) = log10(1 + 1/d)
df['Probabilidade esperada Pe(d)'] = df['Primeiro Dígito'].apply(lambda d: math.log10(1 + 1/d))

# Calcule as frequências esperadas usando a fórmula frequencia_esperada = Pe(d) * n
df['Frequência esperada'] = df['Probabilidade esperada Pe(d)'] * n

# Calcule as probabilidades observadas usando a fórmula probabilidade_observada = frequencia_observada / n
df['Probabilidade observada'] = df['Frequência observada'] / n

# Calcule a estatística z usando a fórmula zcalc = |po - pe| / sqrt(pe * (1 - pe) / n)
df['zcalc'] = df.apply(lambda row: abs(row['Probabilidade observada'] - row['Probabilidade esperada Pe(d)']) / math.sqrt(row['Probabilidade esperada Pe(d)'] * (1 - row['Probabilidade esperada Pe(d)']) / n), axis=1)

# Exiba o DataFrame
display(df)
```
