```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
import statistics
```

```python
df = pd.read_excel('Pasta1.xlsx')
```

```python
df = df.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6'], axis = 1)
```

```python
display(df)
```

```python
correlação = df.corr()
display(correlação)
```

```python
def classificar_correlacao(valor):
    if valor == 1:
        return 'Correlação perfeita'
    elif valor >= 0.9:
        return 'Correlação muito forte'
    elif valor >= 0.7:
        return 'Correlação forte'
    elif valor >= 0.5:
        return 'Correlação moderada'
    elif valor >= 0.3:
        return 'Correlação fraca'
    else:
        return 'Correlação desprezível'
    
for i in range(len(correlação)):
    for j in range(i+1, len(correlação)):  
        valor_correlacao = correlação.iloc[i, j]
        classificacao = classificar_correlacao(valor_correlacao)
        print('Correlação entre o item', i+1, 'e o item', j+1, ':', valor_correlacao, '-', classificacao)
```

# Variancia amostral entre perguntas 




```python
df[['Item 1','Item 2 ','Item 3']].var()
```

```python
pergunta1 = df['Item 1']
pergunta2 = df['Item 2 ']
pergunta3 = df['Item 3']

pergunta1 = np.array(pergunta1)
total = pergunta1.sum()
media = pergunta1.mean()
```

```python
diferença = [(i - media) for i in pergunta1]
quadrado_dif = [(i**2) for i in diferença]
quadrado_dif = np.array(quadrado_dif)
soma_quadrados = quadrado_dif.sum()

variancia_p1 = soma_quadrados / (total - 1)
```

```python
variancia_função = statistics.variance(pergunta1)
variancia_função
```

```python
print(variancia_p1, variancia_função)
```
