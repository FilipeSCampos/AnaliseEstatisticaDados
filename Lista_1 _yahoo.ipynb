```python
import pandas as pd
import yfinance as yf

# Define o período de cinco anos atrás até a data atual
end_date = pd.Timestamp.now()
start_date = end_date - pd.DateOffset(years=5)

# Obtém os dados históricos do dólar com o ticker "USDBRL=X" (dólar para real)
df = yf.download('USDBRL=X', start=start_date, end=end_date)

# Filtra os dados apenas para o mês de agosto
df_august = df[df.index.month == 8]

# Exibe o DataFrame com os dados do dólar em agosto nos últimos 5 anos
display(df_august)
```

```python
df_august['Close'].plot(figsize = (16, 10))
```

```python
df_august['SMA30D'] = df_august['Close'].rolling(window = 30).mean()
df_august
```

```python
df_august['SMA7D'] = df_august['Close'].rolling(window = 7).mean()
df_august
```

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(df_august['Close'], label = 'Dolar')
ax.plot(df_august['SMA30D'], label = 'SMA30')
ax.plot(df_august['Close'].rolling(window = 7).mean(), label = 'SMA7D')

ax.set_title('Dolar', fontsize = 20)
ax.set_ylabel('Preço ($)', fontsize = 16)

plt.legend()
plt.show()
```

```python
# Solicita o número de classes para o histograma
num_classes = int(input("Digite o número de classes para o histograma: "))

# Cria o histograma com o número de classes especificado pelo usuário
plt.hist(df_august['Close'], bins=num_classes, edgecolor='k', color = 'orange')

# Define os rótulos dos eixos
plt.xlabel('Preço de Fechamento do Dólar')
plt.ylabel('Frequência')

# Mostra o histograma
plt.show()
```
