```python
import numpy as np
import pandas as pd
from math import sqrt
```

```python
frequencia = np.array([40,30,10,15,5])
classe = np.array([1,3,5,7,9])
total_amostra = 100
```

```python
media = sum(frequencia*classe)/total_amostra
print('Média',media)

diferenca_quadrada = frequencia *(classe - media)**2
variancia_amostral = np.sum(diferenca_quadrada) / (total_amostra - 1)
variancia_populacional = np.sum(diferenca_quadrada) / total_amostra
desvio_padrao_amostral = sqrt(variancia_amostral)
desvio_padrao_populacional = sqrt(variancia_populacional)
print('Variancia amostral',variancia_amostral)
print('Variancia populacional',variancia_populacional)
print('Desvio padrao amostral',desvio_padrao_amostral)
print('Desvio padrao populacional',desvio_padrao_populacional)
```

```python
legendasfabricantes = ['R', 'C', 'N', 'N', 'I', 'B', 'C', 'C', 'B', 'N', 'R', 'C',
'I', 'C', 'B', 'B', 'C', 'R', 'B', 'I', 'B', 'I', 'R', 'I', 'I', 'R', 'C', 'R', 'N',
'N', 'C', 'C', 'B', 'B', 'R']
# Criar um DataFrame com os dados
df = pd.DataFrame(legendasfabricantes, columns=['Legenda'])
valores = df.value_counts()
percentuais = valores/ len(df) * 100
print(valores, percentuais)
```

```python
# Solicitar a entrada da legenda desejada
legenda_desejada = input("Digite a legenda desejada para consulta: ")
# Verificar se a legenda solicitada está presente no DataFrame
if legenda_desejada in valores.index:
 quantidade_legenda = valores[legenda_desejada]
 percentual = percentuais[legenda_desejada]

 # Imprimir os resultados
 print(f"A legenda especificada possui as seguintes características:")
 print(f"Legenda solicitada: {legenda_desejada}")
 print(f"Quantidade de fabricantes com essa legenda: {quantidade_legenda}")
 print(f"Percentual em relação ao total: {percentual:.2f}%")
else:
 print("Legenda não encontrada no DataFrame.")
```

```python
data = pd.bdate_range(start='2022-08-01', end='2022-08-31', )
cotacao = [5.1606, 5.2329, 5.2846, 5.2409, 5.2165, 5.1247, 5.1224, 5.0497,
5.1127, 5.1023, 5.0925,
 5.134, 5.1784, 5.1773, 5.1961, 5.1709, 5.103, 5.105, 5.1173, 5.0903,
5.0428, 5.0617, 5.179]
data_dia =list(map(int, [d.strftime('%d') for d in data]))
dataframe = {
    'Data': data,
    'Cotacao': cotacao,}
df = pd.DataFrame(dataframe)
#display(df)
#df.info()
variacao_percent = [-0.54]+[(cotacao[i] - cotacao[i - 1])/ cotacao[i - 1] * 100 for i in range(1, len(cotacao))]

df['Variacao%'] = variacao_percent

display(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize = (9,11))
ax = fig.add_subplot(111, projection='3d')

# Plotando o gráfico 3D
ax.plot_trisurf(data_dia, df['Cotacao'], df['Variacao%'], cmap='Purples_r')

# Definindo os rótulos dos eixos
ax.set_xlabel('Data')
ax.set_ylabel('Cotacao')
ax.set_zlabel('Variacao')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotando o gráfico 3D
ax.scatter(data_dia, df['Cotacao'], df['Variacao%'], c='Purple', marker = 'o')

# Definindo os rótulos dos eixos
ax.set_xlabel('Data')
ax.set_ylabel('Cotacao')
ax.set_zlabel('Variacao')

plt.show()
```
